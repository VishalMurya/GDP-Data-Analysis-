### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Dataset

```python
data = pd.read_csv('GDP.csv')
```

```python
df.dropna(inplace = True)
```

### Plot a line plot and compare the gdp growth of India,China,US, UK and Pakistan from 2000-2018

```python
countries = ['India', 'China', 'Pakistan','United States', 'United Kingdom']
```

```python
df[df['Country'].isin(countries)]
```

```python
India = df[df['Country'] == 'India'].values[0]
```

```python
India[1:-2]
```

```python
years = df.columns[1:-2]
years
```

```python
plt.figure(figsize=(10,5))
for col in countries:
    x = df[df['Country'] == col].values[0]
    plt.plot(years,x[1:-2], label =col , marker = 'o' , linestyle = '-' , mec='black')
    plt.title('GDP Compersion between in INDIA , CHINA Pakistan US UK ')
    plt.xlabel('Years')
    plt.ylabel('NO.GDP')
    plt.legend()
    plt.grid()
    plt.xticks(rotation=(45))
```

### Compare the GDP of India and Pakistan with a bar-plot from 2000 - 2018 

```python
country = ['India', 'Pakistan','China']
```

```python
plt.figure(figsize=(10,4))
for col in country:
    y = df[df['Country'] == col].values[0]
    plt.bar(years,y[1:-2], label = col ,alpha = 0.5)
    plt.title('Comparison The GDP od INDIA CHINA PAKISTAN')
    plt.xlabel('Years')
    plt.ylabel('NO. OF GDP')
    plt.legend()
    plt.grid()
    plt.xticks(rotation =(45))
```

###  Get The Top Countries With the Highest Average GDP

```python
df['Average'] = df['Average'].astype('float64')
```

```python
df['Average'].dtypes
```

```python
Coun_tries = df[['Country', 'Average']]
sor_ted  = Coun_tries.sort_values(by='Average', ascending=False)
top_20_countries = sor_ted.head(20)
print("Top 20 Countries with the Highest Average GDP:")
print(top_20_countries)
```

```python
plt.figure(figsize=(10,4))
plt.bar(top_20_countries['Country'], top_20_countries['Average'])
plt.xticks(rotation =(90))
plt.show()
```
